# Gather data

```python
with open("urls_texty.txt", "r") as f:
    urls = json.loads(f.read())
```

```python
data = list()
main_url = "https://texty.org.ua"

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Get title
    try:
        title = soup.find("h1", class_="page-title short-title").text
    except AttributeError:
        try:
            title = soup.find("h1", class_="page-title medium-title").text
        except AttributeError:
            try:
                title = soup.find("h1", class_="page-title long-title").text
            except AttributeError:
                continue


    # Get subtitle
    try:
        subtitle = soup.select("article>section>p")[0].text
    except:
        continue

    # Get text
    text = ""
    for row in soup.select("article>p")[:-2]:
        if row.text[:4]=="Фото":
            pass
        else:
            text += row.text+"|"

    # Get image url
    try:
        image = main_url + soup.select("figure>img")[0].get("data-src")
    except:
        continue

    #Get tags
    tags = list()
    for tag in soup.find_all("a", class_="tag"):
        tags.append(tag.text)
    tags = "|".join(tags)

    pub_date = soup.find("time").text.strip()


    data.append([title, subtitle, text, tags, image, pub_date])
```

```python
df = pd.DataFrame(data, columns=["Title", "Subtitle", "Text", "Tags", "Image", "Pub_date"])
```

```python
df = df.dropna()
```

# Work with data

## Get list with unique tags

```python
raw_df = pd.read_csv("raw_data.csv")
```

```python
all_tags = list()
for tags in raw_df["Tags"]:
    i = 0
    for tag in tags.split("|"):
        # morph parse word
        p_tag = morph.parse(tag)[0]
        nf_tag = p_tag.normal_form
        if nf_tag in all_tags:
            i += 1
            break
    if i > 0:
        continue
    all_tags.append(tags.split("|")[0])
all_tags = list(set(all_tags))
```

```python
# with open("tags_texty.txt", "w") as f:
#     f.write(json.dumps(list(all_tags)))
```

```python
def normalize_tags(tags):
    nf_tags = list()
    for tag in tags.split("|"):
        # morph parse word
        p_tag = morph.parse(tag)[0]
        nf_tag = p_tag.normal_form
        nf_tags.append(nf_tag)
    return nf_tags


def normalize_text(text):
    text = re.sub(r'[^\w\s]', '', text).lower().split()
    normal_forms = list()
    for word in text:
        parsed_word = morph.parse(word)[0]
        normal_form = parsed_word.normal_form
        normal_forms.append(normal_form)
    text = ' '.join(normal_forms)
    return text
```

```python
df = pd.DataFrame({
    "Text": raw_df["Text"].apply(normalize_text),
    "Tags": raw_df["Tags"].apply(normalize_tags),
    "Pub_date": raw_df["Pub_date"]
})
```

# Analize with apriori

```python
import apyori
```

```python
record = list()
for text in df["Text"].values:
    record.append([t for t in text.split(" ") if t in all_tags])
```

```python
tokens = list()

rules = apyori.apriori(record, min_support=0.001, min_confidence=0.1, min_lift=4)
for item in rules:
    items = [x for x in item[0]]
    if len(items) == 2:
        print("Rule: ", items)
        print("support: " + str(item[1]))
        print("Confidence: " + str(item[2][0][2]))
        print("Lift: " + str(item[2][0][3]))
        print("=================================")
        tokens.append([items, item[2][0][3]])
```

```python
tokens = sorted(tokens, key=lambda x: x[1], reverse=True)
```

# Visuals

```python
df2 = pd.DataFrame({
    "Text": [list(set(i)) for i in record],
    "Pub_date": df["Pub_date"].apply(lambda x: x[:7])
})
```

```python
df2
```

```python
by_dates = {tag : {date: 0 for date in reversed(df2["Pub_date"].unique())} for tag in all_tags}

for row in df2.iloc:
    for tag in row[0]:
        by_dates[tag][row[1]] += 1
```

```python
# Most often tags
sum_by_tags = dict()
for tag, dates in by_dates.items():
    sum_by_tags[tag]=sum(dates.values())
MoT = dict(sorted(sum_by_tags.items(), key=lambda x: x[1], reverse=True))
```

```python
chosen_tags = list(MoT.keys())[:5]
```

```python
chosen_tags
```

```python
fig = plt.figure(figsize=(18, 9))
for tag in chosen_tags:
    plt.plot(by_dates[tag].keys(), by_dates[tag].values(), label=tag, linewidth=2.5)
    plt.xticks(rotation=45)
    plt.legend()
```

```python
often = {tag: 0 for tag in all_tags}
for text in df2["Text"]:
    for word in text:
        often[word] += 1
often = dict(sorted(often.items(), key=lambda x: x[1], reverse=True)[:20])

fig = plt.figure(figsize=(18,9))
plt.bar(often.keys(), often.values())
plt.xticks(rotation=45)
plt.xlabel('Words')
plt.ylabel('Частота')
plt.tight_layout()
plt.show()
```
