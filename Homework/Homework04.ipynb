# Завдання 1
	Напишіть програму "Угадай число". Програма загадує деяке число. Користувач повинен його відгадати
	Приклад діалогу має мати наступний вигляд:

		----------------**********----------------
		Введіть своє ім'я:

		>>>> Ігор

		Вгадайте число від 1 до 10. У вас три спроби.
		Введіть число:
		
		>>>> 1

		Помилка. Ваша перша спроба невірна.
		Підказка: задумане число більше за 1. 
		Спробуйте ще раз! 
		Введіть число:

		>>>> 10

		Помилка. Ваша перша спроба невірна.
		Підказка: задумане число менше 10. 
		Спробуйте ще раз! 
		Введіть число:

		>>>> 7

		Помилка. 
		Ігор, гра закінчена.
		
		----------------**********----------------

```python
import random

guess = random.randint(1, 10)
name = input("Введіть своє ім'я: ")
print("Вгадайте число від 1 до 10. У вас три спроби.")

i = 0
while i < 3:
    x = int(input("Введіть число:"))
    if x == guess:
        print("\nВітаю, ви відгадали число!")
        break
    elif x > guess:
        print("\nПомилка. Ваша перша спроба невірна.\nПідказка: задумане число менше", x)
        i += 1
    elif x < guess:
        print("\nПомилка. Ваша перша спроба невірна.\nПідказка: задумане число більше", x)
        i += 1
    elif i == 2:
        print(f"\nПомилка.\n{name}, гра закінчена.")
```

# Завдання 2
	Напишіть програму "Митниця". Алгоритм її роботи наступний: програма на вхід получає імена людей, що перетинають кордон. 
	Ввід закінчується тоді, коли введено слово "END"
	При завершені роботи, програма виводить ім’я людини та кількість перетинів кордону. 

		----------------**********----------------
		Хто перетинає кордон:

		>>>> Ігор

		Хто перетинає кордон:

		>>>> Оля

		Хто перетинає кордон:

		>>>> Ігор		

		Хто перетинає кордон:

		>>>> Ігор		
		
		Хто перетинає кордон:

		>>>> Андрій

		Хто перетинає кордон:

		>>>> END
		
		Ігор - 3
		Оля - 1
		Андрій - 1
		
		----------------**********----------------

```python
dic = {}

while True:
    name = input("Хто перетинає кордон:  ")
    if name == "END":
        break
    elif name not in dic:
        dic[name] = 1
    elif name in dic:
        dic[name] += 1

for key, value in dic.items():
    print(f"{key} - {value}")
```

# Завдання 3
	Створіть матрицю чисел 3х4. Заповніть її цілими числами від 0 до 10. Знайдіть:
		- Суму всіх чисел матриці.
		- Максимальний елемент.
		- Суму чисел в першому рядку.
		- Мінімальний елемент в  другому стовбці.
        - Максимальний елемент по головній діагоналі (діагональ, що проходить через лівий верхній та правий нижній кути).

```python
# Задаємо матрицю
import random

l = [[0 for j in range(4)] for i in range(3)]

for i in range(3):
    for j in range(4):
        l[i][j] = random.randint(1, 10)
```

```python
# How matrix looks like
for i in l:
    print(i)
```

```python
su = 0
su1 = 0
ma = -999999999
mi2 = 999999999
maxd = -999999999

for i in range(3):
    for j in range(4):
        su += l[i][j]
        
        if l[i][j] > ma:
            ma = l[i][j]
        
        if i == 0:
            su1 += l[i][j]
            
        if j == 1:
            if l[i][j] < mi2:
                mi2 = l[i][j]
        if i == j:
            if l[i][j] > maxd:
                maxd = l[i][j]

print("Сума всіх чисел матриці ", su)
print("Максимальний елемент ", ma)
print("Сума чисел в першому рядку ", su1)
print("Мінімальний елемент в  другому стовбці ", mi2)
print("Максимальний елемент по головній діагоналі ", maxd)
```
