Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
cond = False

while cond == False:
    sex = input("Введіть стать спортсмена: ")
    age = int(input("Введіть вік спортсмена: "))
    cond = True
    if (sex == "f" and age in range(19,101)) or (sex == "m" and age in range(16,101)):
        print("Cпортсмен може брати участь")
    elif (sex != "f" and sex != "m") or (age not in range(0,101)):
        print("Дані введені неправильно, спробуйте ще раз")
        cond = False
    else:
        print("Cпортсмен не може брати участь")
```

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
def min_n_max(l):
    print(f"Максимальне значення - {max(l)}, Мінімальне значення - {min(l)}\n")

def medium(l):
    global medium    #для використання в функції stand_deviation()
    sum = 0
    for i in l:
        sum = sum + i
    medium = sum / len(l)
    print(f"Середнє значення послідовності - {round(medium, 3)}\n")
    
def median(l):
    l.sort()
    if len(l) % 2 == 1:
        print(f"Медіана - {l[int(len(l)/2+1)]}")
    else:
        print(f"Медіана - {(l[int(len(l)/2)]+l[int(len(l)/2+1)])/2}\n")
        
def freq(l):
    for i in range(0,11):
        print(f"Кількість значень {i} - {l.count(i)}")
    print('\n')

def stand_deviation(l):
    sum = 0
    for i in range(len(l)):
        sum = sum + (l[i]-medium)**2
    sig = m.sqrt(1/len(l)*sum)
    print(f"Стандартне відхилення - {round(sig, 3)}")

    

import random
import math as m

#генеруэмо тисячу чисел в діапазоні від 0 до 10
l = [random.randint(0,10) for i in range(1000)]

#Використовуєм функції
min_n_max(l)
medium(l)
median(l)
freq(l)
stand_deviation(l)
```
