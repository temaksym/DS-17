2. Отримати курс долара за останній рік.

```python
import requests
import datetime
import pickle

dic = {}
dt = datetime.datetime(2021, 1, 1)

for i in range(365):
    url = "https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?date="+ (dt + datetime.timedelta(days=i)).strftime("%Y%m%d") + "&json"
    data = requests.get(url).json()
    for j in data:
        if j["cc"] == "USD":
            dic[(dt + datetime.timedelta(days=i)).strftime("%Y-%m-%d")] = j["rate"]
```

```python
dic
```

3. Вивести середнє значення та відхилення курсу за кожний місяць.

```python
su = 0
quantity = 0
li = []
info = ""

for i in range(1,13):
    for date, val in dic.items():
        if int(date[5:7]) == i:
            su += val
            quantity += 1
            li.append(val)
    print("Місяць", i, ": середнє значення -", round(su/quantity, 2),"    відхилення ", round(li[-1]-li[1], 2))
    info += "Місяць " + str(i) + " : середнє значення - " + str(round(su/quantity, 2)) + "    відхилення " + str(round(li[-1]-li[1], 2)) + "\n"
    li = []
```

```python
print(pd.read_sql_query('''SELECT pr.prod_name, sum(oi.quantity) as "Quantity", oi.item_price
                            FROM OrderItems oi
                            JOIN Products pr
                            ON pr.prod_id == oi.prod_id
                            GROUP BY oi.prod_id
                            ORDER BY 2 DESC, 3 DESC;''', conn))
```

14. Виведіть список клієнтів та їх замовлення (використайте JOIN)

```python
print(pd.read_sql_query('''SELECT prod_name, prod_price
                            FROM Products pr
                            WHERE prod_price < 5
                            UNION
                            SELECT prod_name, prod_price
                            FROM Products pr
                            WHERE prod_price >= 5
                            ORDER BY 2 DESC;''', conn))
```
