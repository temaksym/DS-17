```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, "navy")
plt.ylabel("Y")
plt.xlabel("X")
plt.title("First")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "g-*")

plt.subplot(1, 2, 2)
plt.plot(x, y, "r-")
```

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x, x**2)
ax[0][1].plot(x, x**(1/2),"navy")
ax[1][0].plot(x, 1/(x+1),"b-^")
ax[1][1].plot(x, x**4,"purple")

plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.fill_between(x, x**2, x**3, )
```
