```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку

 - останній це пряма пропорційність

```python
plt.figure(figsize=(15,8))
plt.subplot(221)
plt.plot(df['a'])
plt.subplot(222)
plt.plot(df['b'])
plt.subplot(223)
plt.plot(df['c'])
plt.subplot(224)
plt.plot(np.linspace(1,30))
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
fig, ax = plt.subplots(3,1, figsize=(18,10))

ax[0].plot(df["a"], label="Line A")
ax[0].text(-1,35,"Line A", fontsize=15, color="navy")
ax[1].plot(df["b"], label="Line B")
ax[1].text(-1,35,"Line B", fontsize=15, color="navy")
ax[2].plot(df["c"], label="Line C")
ax[2].text(-1,35,"Line C", fontsize=15, color="navy")
```

Створити візуалізацію, аналогічно рисунку

```python
plt.figure(figsize=(20,10))
plt.bar(range(len(df['a'])), df['a'], color='tomato', alpha=0.5)
```
