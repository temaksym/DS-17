- ведіть топ-3 продукти, що продаються
- виведіть топ-2 пари продуктів, що продаються

```python
import pandas as pd
```

```python
df = pd.read_csv("Groceries.csv")
df
```

```python
df.describe().transpose()[["mean", "max", "75%", "50%", "25%"]]
```

```python
import apyori
```

```python
record = list()
for row in df.drop("Item(s)", axis=1).values:
    record.append([str(it) for it in row if not pd.isna(it)])
# list of all receipts
```

## Топ-2 продукти що продаються

```python
rules = apyori.apriori(record, min_support=0.0045, min_confidence=0.2, min_lift=3.8)
for item in rules:
    items = [x for x in item[0]]
    if len(items) == 2:
        print("Rule: ", items)
        print("support: " + str(item[1]))
        print("Confidence: " + str(item[2][0][2]))
        print("Lift: " + str(item[2][0][3]))
        print("=================================")
```

## Топ-3 продукти що продаються

```python
rules = apyori.apriori(record, min_support=0.0045, min_confidence=0.2, min_lift=3.9)
for item in rules:
    items = [x for x in item[0]]
    if len(items) == 3:
        print("Rule: ", items)
        print("support: " + str(item[1]))
        print("Confidence: " + str(item[2][0][2]))
        print("Lift: " + str(item[2][0][3]))
        print("=================================")
```

Я вважаю що було б доречно використати цю інформацію:
- для того аби визначати як краще розміщувати товари (товари з високим Lift - поруч)
- для рекомендацій (наприклад при доставці продуктів в пропозиціях "разом вигідніше")

- Опишіть отримані результати

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

# Read the data into a pandas DataFrame
data = pd.read_csv('Mall_Customers.csv')
data
```

```python
import matplotlib.pyplot as plt

label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])

X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Determine how many clusters we need
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)


plt.plot(range(1, 11), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
n_clusters = 5

kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
kmeans.fit(X)

data['Cluster'] = kmeans.labels_

print(data[['CustomerID', 'Cluster']])
```

```python
import seaborn as sns
sns.set_theme()
# fig, axes = plt.subplots(1,2, figsize=(18,9))
sns.lmplot(data=data, y="Spending Score (1-100)", x="Age", hue="Cluster")
sns.lmplot(data=data, y="Spending Score (1-100)", x="Annual Income (k$)", hue="Cluster")
```

```python
for i in range(5):
    print("\nCategory ", i, "--------------------------------------------------------")
    print(data[data['Cluster']==i].drop(["CustomerID", "Cluster"], axis=1).describe())
```
