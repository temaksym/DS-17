# Using GeoPY

```python
from geopy.geocoders import ArcGIS
from geopy.distance import geodesic
```

```python
st_addresses = ['Академмістечко','Житомирська','Cвятошин','Нивки','Берестейська','Шулявська','Політехнічний інститут','Вокзальна','Університет','Театральна','Хрещатик','Дніпро','Арсенальна','Гідропарк','Лівобережна','Дарниця','Чернігівська','Лісова','Героїв Дніпра','Мінська','Почайна','Оболонь','Тараса Шевченка','Контрактова площа','Поштова площа','Майдан Незалежності','Площа Льва Толстого','Олімпійська','Палац «Україна»','Либідська','Деміївська','Голосіївська','Васильківська','Виставковий центр','Іподром','Теремки','Сирець','Дорогожичі','Лук\'янівська','Золоті ворота','Палац спорту','Кловська','Печерська','Дружби народів','Видубичі','Славутич','Осокорки','Позняки','Харківська','Вирлиця','Бориспільська','Червоний хутір']
st_coor = list()
for st_address in st_addresses:
    st_loc = ArcGIS().geocode("Київ, станція метро "+st_address)
    st_coor.append((st_loc.latitude, st_loc.longitude))
# Отримали список координат всіх станцій
```

Додаємо стовпці координа квартири і відстані до найближчого метро

```python
data["Координати"] = None
data["Відстань"] = None
for i in range(len(data)):
    # Знаходимо координати квартири
    loc = ArcGIS().geocode("Київ, "+data["Адреса"][i])
    data["Координати"][i] = (loc.latitude, loc.longitude)
    
    # Рахуємо відстань до найбдижчого метро
    dist = list()
    for coor in st_coor:
        dist.append(geodesic(data["Координати"][i], coor))
    data["Відстань"][i] = int(float(str(min(dist))[:-3])*1000)
```

# Підготовка даних до використання

```python
# Видаляємо всі рядки які не мають змісту
data = data.drop_duplicates()
data = data.drop(data[data["Відстань"]==0.0].index)
data = data.drop(data[data["Відстань"]>5000].index)
data = data.drop(data[data["Ціна, грн"]>50000].index)
data = data.drop(data[data["Кімнати"]>6].index)
data = data.drop(data[data["Площа"]>200].index)
data = data.drop(data[data["Район"].isin(["Куренівка","Чорна Гора","Звіринець","Солом'янка","Сирець"])].index)
```

```python
distr = dict()
for district in districts:
    distr[district] = np.median(np.array(list(data[data["Район"]==district]["Ціна, грн"])))
distr = dict(sorted(distr.items(), key=lambda x: x[1]))
distr
```

```python
distr_keys = dict()
for district, median_v in distr.items():
    if median_v >= 12000:
        distr_keys[district] = 3
    elif median_v >= 10000:
        distr_keys[district] = 2
    elif median_v > 0:
        distr_keys[district] = 1
    else:
        raise ValueError
distr_keys
```

```python
data["Рейтинг"] = data["Район"].map(distr_keys)
data
```

```python
data.to_csv("flats.csv", index=False)
```

# Machine Learning

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
reg = LinearRegression()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20,8))

y = np.array(list(data["Ціна, грн"])).reshape(-1, 1)

X = np.array(list(data["Відстань"])).reshape(-1, 1)
reg.fit(X, y)
x_pred = np.linspace(0, 5000, 100).reshape(-1, 1)
y_pred = reg.predict(x_pred)
axes[0].plot(x_pred, y_pred)
axes[0].text(0.5, 1, "Відстань до метро")

X = np.array(list(data["Кімнати"])).reshape(-1, 1)
reg.fit(X, y)
x_pred = np.linspace(1, 5, 5).reshape(-1, 1)
y_pred = reg.predict(x_pred)
axes[1].plot(x_pred, y_pred)
axes[1].text(0.5, 1, "К-сть кімнат")

X = np.array(list(data["Площа"])).reshape(-1, 1)
reg.fit(X, y)
x_pred = np.linspace(0, 175, 25).reshape(-1, 1)
y_pred = reg.predict(x_pred)
axes[2].plot(x_pred, y_pred)
axes[2].text(0.5, 1, "Площа")

X = np.array(list(data["Рейтинг"])).reshape(-1, 1)
reg.fit(X, y)
x_pred = np.linspace(0, 3, 3).reshape(-1, 1)
y_pred = reg.predict(x_pred)
axes[3].plot(x_pred, y_pred)
axes[3].text(0.5, 1, "Рейтинг")
''
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['Площа', 'Кімнати', 'Відстань', 'Рейтинг']], data['Ціна, грн'], test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print('Mean squared error:', mse)
```
